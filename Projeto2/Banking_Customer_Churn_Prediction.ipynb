# Data Preparation

In this section, we focus on preparing the dataset for model training and evaluation by separating it into training and testing subsets. This step is crucial for assessing the model's performance on unseen data and ensuring its generalizability.

1. **Categorical and Continuous Features**: 
   - The dataset includes both categorical and continuous features. Categorical features such as 'HasCrCard', 'IsActiveMember', 'Geography', and 'Gender' are discrete variables that classify data into specific categories.
   - Continuous features, including 'CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', and 'EstimatedSalary', represent numerical values that can take on a range of values.

2. **Data Splitting**:
   - The data is split into training and testing subsets using a 75-25 split ratio. This ensures that 75% of the data is used for training the model, while the remaining 25% is reserved for testing the model's performance.
   - The splitting is performed with a fixed random state (100) to ensure reproducibility of results.

3. **Feature Selection**:
   - The training data is constructed by combining the target variable 'Exited' with both continuous and categorical features. This comprehensive selection ensures that the model is exposed to all relevant information for making accurate predictions.

By carefully preparing the data in this manner, we set a strong foundation for building robust machine learning models that can effectively predict customer churn.

```python
# Separating into trainning and testing data
categorical = ['HasCrCard', 'IsActiveMember','Geography', 'Gender']
continuous = ['CreditScore',  'Age', 'Tenure', 'Balance','NumOfProducts', 'EstimatedSalary']

train = churn_data.sample(frac=0.8, random_state=100)
test = churn_data.drop(train.index)
train = train[['Exited'] + continuous + categorical]
train.head()
```

#### Encoding Categorical Variables

Categorical variables must be encoded into numerical values to be processed by machine learning algorithms. We use Label Encoding to transform these variables. Label Encoding assigns a unique integer to each category of the variable, which, although not maintaining any ordinal relationship, allows for the inclusion of these categorical features in our models.

1. **Process**:
   - **Initialization**: Create a `LabelEncoder` instance.
   - **Transformation**: Encode each categorical variable, converting categories to numerical values.

2. **Application**:
   - Apply the encoder to each categorical column.
   - Ensure all categorical variables are now numerical, making the data suitable for machine learning.

3. **Verification**:
   - Verify the transformed data by checking a subset, ensuring all categorical variables are numerically encoded.

This step ensures compatibility with machine learning models, facilitating accurate and efficient model training.


```python
# Encoding categorical variables using Label Encoder
le = LabelEncoder()
for column in categorical:
    churn_data[column] = le.fit_transform(churn_data[column])
    
train.head()    
```

#### Normalizing Numerical Features

Numerical features are often on different scales, which can bias a model towards variables with larger magnitudes. To prevent this, we apply StandardScaler to normalize these features, giving them equal weight in the predictive model.

1. **Process**:

    - **Initialization**:
        - Create a `StandardScaler` instance.

    - **Transformation**:
        - Select numerical columns to scale: `['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']`.
        - Apply the scaler to these columns, transforming them to a standard scale with a mean of 0 and a standard deviation of 1.

2. **Application**:

    - Apply the scaler to each numerical column.
    - Ensure all selected numerical variables are now on a comparable scale.

3. **Verification**:

    - Verify the transformed data by checking a subset, ensuring all numerical variables are properly scaled.

This step eliminates bias towards features with larger magnitudes, ensuring a balanced contribution from all features, which enhances the performance and accuracy of the predictive model.


```python
# Standardizing numerical columns with StandardScaler
scaler = StandardScaler()
numerical_columns = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']
churn_data[numerical_columns] = scaler.fit_transform(churn_data[numerical_columns])

train.head() 
```

```python
# Function to split data into training and testing sets
def get_data():
    # Select 80% of the data for training randomly
    train = churn_data.sample(frac=0.8, random_state=100)
    # Select the remaining data for testing
    test = churn_data.drop(train.index)
    return train, test

# Function to prepare data for modeling
def prepare_data(data):
    # Separate features (X) and target variable (y)
    X = data.drop('Exited', axis=1)
    y = data['Exited']
    return X, y

# Function to evaluate models using F1 score
def evaluate_models1(models, X, y):
    results = []
    algos = []
    
    for algo, model in models:
        # Define KFold cross-validation with 90 splits
        kfold = KFold(n_splits=90, random_state=30, shuffle=True)
        # Calculate F1 score for each model
        cv_score = cross_val_score(model, X, y, cv=kfold, scoring='f1')
        results.append(cv_score)
        algos.append(algo)
        
    return results, algos

# Function to evaluate models using accuracy score
def evaluate_models2(models, X, y):
    results = []
    algos = []
    
    for algo, model in models:
        # Define KFold cross-validation with 90 splits
        kfold = KFold(n_splits=90, random_state=30, shuffle=True)
        # Calculate accuracy score for each model
        cv_score = cross_val_score(model, X, y, cv=kfold, scoring='accuracy')
        results.append(cv_score)
        algos.append(algo)
        
    return results, algos

# Function to plot F1 score results
def plot_results1(results, algos):
    colors = ['skyblue']
    plt.figure(figsize=(10,10))
    # Create boxplot for F1 scores
    box = plt.boxplot(results, labels=algos, patch_artist=True)
    plt.title('Algorithm F1 Score Comparison')
    plt.xlabel('Algorithm')
    plt.ylabel('F1 score')
    plt.xticks(rotation=90)

    # Set colors for the boxes
    for patch, color in zip(box['boxes'], colors):
        patch.set_facecolor(color)

    plt.show()

# Function to plot accuracy score results
def plot_results2(results, algos):
    colors = ['skyblue']
    plt.figure(figsize=(10,10))
    # Create boxplot for accuracy scores
    box = plt.boxplot(results, labels=algos, patch_artist=True)
    plt.title('Algorithm Accuracy Comparison')
    plt.xlabel('Algorithm')
    plt.ylabel('Accuracy')
    plt.xticks(rotation=90)

    # Set colors for the boxes
    for patch, color in zip(box['boxes'], colors):
        patch.set_facecolor(color)

    plt.show()
    
# Function to find the best models based on F1 score and accuracy
def best_models():
    # Get training and testing data
    train, _ = get_data()
    X_train, y_train = prepare_data(train)
    
    # Define a list of models to evaluate
    models = [
        ('SVM', SVC()),
        ('Logistic Regression', LogisticRegression()),
        ('Decision Tree', DecisionTreeClassifier()),
        ('Extra Trees', ExtraTreesClassifier()),
        ('Quadratic Discriminant Analysis', QuadraticDiscriminantAnalysis()),
        ('CatBoost', CatBoostClassifier(logging_level='Silent', random_state=30)),
        ('Random Forest', RandomForestClassifier()),  
        ('Linear Discriminant Analysis', LinearDiscriminantAnalysis()),       
        ('MLP', MLPClassifier(max_iter=1000)), 
        ('KNN', KNeighborsClassifier()),
        ('Naive Bayes', GaussianNB()),
        ('AdaBoost', AdaBoostClassifier()),
        ('Bagging', BaggingClassifier()),
        ('Gradient Boosting', GradientBoostingClassifier()),
        ('XGBoost', XGBClassifier()),
    ]

    # Evaluate models using F1 score and accuracy
    results1, algos1 = evaluate_models1(models, X_train, y_train)
    results2, algos2 = evaluate_models2(models, X_train, y_train)
    
    # Plot results
    plot_results1(results1, algos1)
    plot_results2(results2, algos2)

# Call the function to find and plot the best models
best_models()
```
